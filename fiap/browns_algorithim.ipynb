```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import time
```

```python
med = 3 # media assumida para simução
N = 10000 # número de simulações
std = 0.03 # desvio padrão
T = 250 # tempo de simulação
dt = 1/T # intervalo de tempo
```

```python
S = np.zeros((T, N))
S[0] = 100
```

#### Simulações

```python
start = time.time()
for t in range(1,T):
  Z = np.random.standard_normal(N)
  S[t] = S[t-1]*np.exp(med - (std**2/2)*dt + std*((dt)**0.5)*(Z))

stop = time.time()
print(f"Tempo de execução: {(stop-start):.2f} s")
```

#### Plotar Figura


```python
plt.figure(figsize=(10,4))
plt.plot(S[0:180, 0:100])
plt.xlabel('Tempo')
plt.ylabel('Valor previsto')
plt.title('Simulação de preço de ação')
plt.grid(True)
plt.show()
```
