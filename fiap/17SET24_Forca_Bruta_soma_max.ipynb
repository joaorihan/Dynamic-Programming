#### Este Notebook versa sobre Força Bruta para resolver desafio de se encontrar a maxima soma (subsequente) em um array.

```python
# Importação de Bibliotecas
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Definição de Parâmetros Técnicos
NN = 250 # comprimento máximo dos arrays a serem testados
N = 10 # comprimento do array_exemplo
array = np.random.randint(-10,10,N)
print(array)
```

```python
# Definição de Função
def max_soma_forca_bruta(array):
  # array = input a ser fornecido
  soma_local = 0 # contador local da soma no interior do array
  max_soma = 0 # contador geral para receber a maior soma apurada
  n = len(array) # tamanho do array
  for i in range(n): # loop para percorrer o array - referencia = i
    for j in range(i, n): # loop começando da referencia
      soma_local = sum(array[i:j+1]) # soma local do array
      max_soma = max(max_soma, soma_local) # separar a maior soma apurada
  return max_soma
```

```python
# Testes
array = np.random.randint(-10,10,N)
start = time.time()
v = max_soma_forca_bruta(array)
stop = time.time()
sprint = stop - start
sprint = round(sprint,6)
print('array =',array)
print('max_soma =',v,'tempo_execucao =',sprint)
```

#### Exploracao da dificuldade computacional

```python
dificult = pd.DataFrame() # dataframe auxiliar
tempo = pd.Series() # estrutura de dados auxliar
tamanho = pd.Series() # estrutura de dados auxiliar
for n in range (2,NN): # comecar com array de 2 elementos no minimo e no maximo NN
  array = np.random.randint(-10,10,n)
  start = time.time()
  v = max_soma_forca_bruta(array)
  stop = time.time()
  sprint = stop - start
  sprint = round(sprint,6)
  tamanho = np.append(tamanho,n)
  tempo = np.append(tempo,sprint)
dificult['tamanho'] = tamanho
dificult['Tempo_execucao'] = tempo
```

```python
plt.figure(figsize=(10,6)) # definição do tamanho da figura
plt.style.use('fivethirtyeight') # estilo de impressao = melhorar visualizacao
sns.scatterplot(data = dificult, x = 'tamanho', y = 'Tempo_execucao', color = 'blue') # grafico com os pontos
plt.xlabel('numero de elementos no array')
plt.ylabel('Tempo de execução - segundos')
plt.title('Tempo de execução em função do tamanho do array')
plt.grid(True)
plt.show()
```
